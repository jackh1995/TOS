```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('FFF_10_100.txt', 'r') as f:
    FFF_10_100 = [np.array(x.split(), dtype=int) for x in f.read().splitlines()]
with open('TFF_10_100.txt', 'r') as f:
    TFF_10_100 = [np.array(x.split(), dtype=int) for x in f.read().splitlines()]
with open('FFF_11_100.txt', 'r') as f:
    FFF_11_100 = [np.array(x.split(), dtype=int) for x in f.read().splitlines()]
with open('TFF_11_100.txt', 'r') as f:
    TFF_11_100 = [np.array(x.split(), dtype=int) for x in f.read().splitlines()]
```

```python
# for t in [FFF_10_100, TFF_10_100, FFF_11_100, TFF_11_100]:
for t in [FFF_11_100, TFF_11_100]:

    plt.figure(figsize=(8, 6))
    for i, x in enumerate(t):
        print(x.mean(), x.std())
        plt.bar(np.arange(len(x))*3+3*i, x, label=i)
    plt.legend()
    plt.show()
    plt.close()
```
