```python
import itertools
import seaborn as sns; sns.set_theme()
def hamming_dist(a, b, ell):
    a_num = [int(x) for x in a]
    b_num = [int(x) for x in b]
    s = sum(np.logical_xor(a_num, b_num))
    return 1 - s/ell
```

```python
sim_list = []

ell = 8
bit_strings = ["".join(x) for x in itertools.product(['0', '1'], repeat=ell)]

n_opt = 4
opt_list = ["".join(x) for x in itertools.product(['0'*n_opt, '1'*n_opt], repeat=ell//n_opt)]

assert ell % n_opt == 0

print(len(bit_strings))
print(len(opt_list))

r = np.empty((len(bit_strings), len(bit_strings)))
for i in range(len(bit_strings)):
    print(f'[{i}]', end='\t')
    
    tmp_list = []
    for x in opt_list:
        tmp = hamming_dist(bit_strings[i], x, ell)
        tmp_list.append(tmp)
    m = max(tmp_list)

    if m == 1:
        print(m, end='\t+\n')
    else:
        print(m)
```
