```python
import random
import numpy as np
```

```python
length = 400
remained_length = 400

bb_list = []

max_bb_size = 7
min_bb_size = 3

def trap (unitary, fHigh, fLow, trapK):
    if unitary > trapK:
        return 0
    if (unitary == trapK):
        return fHigh
    else:
        return fLow - unitary * fLow / (trapK-1)

trap(1, 1, 1-1/5, 5)

while (remained_length):
    if (remained_length >= max_bb_size):
        bb_size = random.randint(min_bb_size, max_bb_size)
        remained_length -= bb_size
    else:
        bb_size = remained_length
        remained_length = 0
    bb_list.append(bb_size)
bb_list
```
